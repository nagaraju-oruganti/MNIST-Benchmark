```python
import numpy as np
from helper_config import Config
from helper_trainer import train

for arch in ['ann', 'annlarge', 'snn']:
  print(f'Selected {arch} for training')
  acc = []
  for fold in range(5):
    config = Config()
    config.fold = fold
    config.data_dir = '../inputs/mnist'
    config.models_dir = '../models'
    config.model_name = f'{arch}_baseline'
    config.train_batch_size = 64
    config.valid_batch_size = 128
    config.iters_to_accumlate = 1
    config.learning_rate = 1e-4
    config.num_epochs = 1000

    config.save_epoch_wait = 1
    config.early_stop_count = 20
    config.save_checkpoint = False

    config.time_steps = 3
    config.train_arch = arch.upper()

    results = train(config)
    val_acc = max([r[-2] for r in results])
    acc.append(val_acc)
    print(f'Best validation accuracy (fold {fold}): {val_acc:.5f}')

  print('Summary')
  print(f'Mean: {np.mean(acc):.5f} | std. dev.: {np.std(acc):.5f}')
```
